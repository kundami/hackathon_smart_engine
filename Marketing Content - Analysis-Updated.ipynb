```python
import glob
from datetime import datetime
import spacy
import pprint
from nltk.util import ngrams
from nltk.stem import PorterStemmer
from gensim import corpora
from gensim import models

import logging
#log_file_name = datetime.now().strftime('log/topic_modelling_%Y-%m-%d_%H-%M-%S.log')
#logging.basicConfig(filename=log_file_name, format='%(asctime)s : %(levelname)s : %(message)s', level=logging.DEBUG)

nlp_en = spacy.load('en_core_web_sm')
```

```python
def preprocess_text(doc1):
    doc = nlp_en(doc1)
    #print("Doc - Orig:", doc)
    # for token in doc:
    #    print(token.text, token.is_stop, token.lemma_)

    # Lemmatize, Remove Stop-Words/Punctuations/Numbers/Symbols and 1-char words
    # ps = PorterStemmer()  --- removed Stemming
    doc = [token.lemma_.lower() for token in doc if token.lemma_ != "-PRON-" and token.is_stop is not True and token.pos_ not in ('PUNCT', 'NUM', 'SYM') and len(str(token)) != 1]
    print("Processed Doc (Lemmatize, Remove Stop-Words/Punctuations/Numbers/Symbols and 1-char words)\n", str(doc) + "\n")

    # (Incomplete)  Add n-grams so terms like "investment management" are taken together by the model
    #    bigram = list(ngrams(doc, 2))
    #    print("bigram:", bigram)
    #    trigram = list(ngrams(doc, 3))
    #    print("\ntrigram:", trigram)

    return doc

```

```python
def train_lda_model(refresh_dictionary):

    num_topics = 10
    chunksize = 200
    passes = 100
    iterations = 300
    num_words = topn = 4
    #update_every = "not added"
    dict_no_below = "not added"
    dict_no_above = "not added"

    # Documents to Train
    content = ["As environmental, social and governance (ESG) factors become more broadly incorporated into investment decision-making, sovereign fixed income investors are grappling with the best way to benchmark and quantify these factors across a diverse group of countries. While some may use a simple approach of investing in countries with the highest third-party ESG ratings, we disagree. Instead, we think that this “paint-by-numbers” approach is systematically biased against developing countries, where the potential for ESG improvement is arguably greater, and would result in a pool of low yielding, highly correlated, and geographically concentrated countries. Morgan Stanley Investment Management has developed what we consider to be a better approach.",
               "Plastic is high on the corporate agenda. We reviewed our holdings to evaluate how this topical issue might influence the sustainability of company returns on operating capital, which is how we assess material risks and opportunities on any environmental, social and governance (ESG) issue. We believe that by addressing their plastic consumption, companies not only benefit the environment, but also create opportunities to differentiate themselves."]

    corpus = []
    for doc in content:
        corpus.append(preprocess_text(doc))

    if refresh_dictionary is True:
        dictionary = corpora.Dictionary(corpus)
        #dictionary.filter_extremes(no_below=dict_no_below, no_above=dict_no_above)
    else:
        dictionary = corpora.Dictionary.load('model/model_dictionary_2019-11-20_18-19-04')

    print(str(datetime.now()) + "\n")
    print(str(dictionary))
    print("Dictionary Tokens:")
    pprint.pprint(dictionary.token2id)
    print("\nCorpus:\n" + str(corpus))

    vect = [dictionary.doc2bow(doc) for doc in corpus]
    print("\nVector:")
    pprint.pprint(vect)

    lda_model = models.ldamodel.LdaModel(corpus=vect,
                                         id2word=dictionary,
                                         num_topics=num_topics,
                                         chunksize=chunksize,
                                         passes=passes,
                                         iterations=iterations,
                                         random_state=2,
                                         eval_every=1,
                                         alpha='auto',
                                         eta='auto')
    #model_file_name = datetime.now().strftime('model/lda_model')
    #lda_model.save(model_file_name)

    print(str(datetime.now()))
    print("\nPARAMETERS to the MODEL:\nnum_topics:" + str(num_topics) + "\n" + "chunksize:" + str(chunksize) + "\n" + "passes:" + str(passes) + "\n" + "iterations:" + str(iterations) + "\n" + "num_words:" + str(num_words) + "\n" + "dict_no_below:" + str(dict_no_below) + "\n" + "dict_no_above:" + str(dict_no_above))
    print("\nTopics listing:")
    pprint.pprint(lda_model.print_topics(num_words=num_words))
    print("\nTopics listing with coherence:")
    pprint.pprint(lda_model.top_topics(corpus=vect, topn=topn))
    print("\nMODEL TRAINING COMPLETED\n")

    print(str(datetime.now()))
    print("\nDocument Inventory Path:")


    # Documents to the Evaluate
    content = ["As environmental, social and governance (ESG) factors become more broadly incorporated into investment decision-making, sovereign fixed income investors are grappling with the best way to benchmark and quantify these factors across a diverse group of countries. While some may use a simple approach of investing in countries with the highest third-party ESG ratings, we disagree. Instead, we think that this “paint-by-numbers” approach is systematically biased against developing countries, where the potential for ESG improvement is arguably greater, and would result in a pool of low yielding, highly correlated, and geographically concentrated countries. Morgan Stanley Investment Management has developed what we consider to be a better approach.",
               "Plastic is high on the corporate agenda. We reviewed our holdings to evaluate how this topical issue might influence the sustainability of company returns on operating capital, which is how we assess material risks and opportunities on any environmental, social and governance (ESG) issue. We believe that by addressing their plastic consumption, companies not only benefit the environment, but also create opportunities to differentiate themselves."]
    index = 1
    for text in content:
        doc = (preprocess_text(text))
        #print("Article Content - Procesed:", doc);

        vect = [dictionary.doc2bow(doc)]
        #print("run_lda_model:vector:")
        #pprint.pprint(vect)

        doc_topics = lda_model.get_document_topics(vect)
        for topic in doc_topics:
            #print(file, ":", topic)
            print("Text"+str(index)+":"+str(topic))
        index = index+1
```

```python
train_lda_model(True)
```
